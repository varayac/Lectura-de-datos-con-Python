# Base de datos propitas

* Objeto a estudiar: cantidad de propinas.
    - `tips` : Cantidad de propinas. Float.
* Atributos de nuestra base de datos:
    - `sex`: Sexo del camarero/a. String.
    - `total_bill` : Total cuenta a consumo.
    - `smoker` : Cantidad fumadores.
    - `day` : Dia de atención.
    - `time` : Almuerzo o Cena.
    - `size` : Cantidad de personas por mesa.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as statm
import statsmodels.formula.api as frm
plt.style.use('seaborn')
```
